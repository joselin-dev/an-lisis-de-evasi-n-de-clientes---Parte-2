## 🎯 Split

```python
X = datos.drop("Churn", axis=1)
y = datos["Churn"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
X_train = pd.get_dummies(X_train, drop_first=True)

X_test = pd.get_dummies(X_test, drop_first=True)

X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)
```

##Entrenamiento

###MODELO 1 -  Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print("Random Forest")
print(classification_report(y_test, y_pred_rf))
```

El modelo es muy bueno predictivos a los clientes que se quedan (fila 0, F1-score de 0.86), pero es algo débil prediciendo a los que se van.

### MODELO 2 - Regresion lineal

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
log_model.fit(X_train, y_train)

y_pred_log_normal = log_model.predict(X_test)

print("Regresion logistica - sin balancear")
print(classification_report(y_test, y_pred_log_normal))
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000, class_weight="balanced")
log_model.fit(X_train, y_train)

y_pred_log = log_model.predict(X_test)

print(classification_report(y_test, y_pred_log))
```

Al efectuar el balanceo cambió radicalmente el rendimiento de modelo de Regresión Logística

##Informe de Análisis de Cancelación de Clientes (Churn) - TelecomX2

El objetivo de este análisis es identificar por qué los clientes abandonan el servicio y proponer acciones concretas para mejorar la retención. Tras evaluar el comportamiento de los clientes y probar diversos modelos, se concluye que la antigüedad, el costo mensual y la estabilidad del cliente son los pilares fundamentales que determinan la fuga.

###Factores Clave que Influyen en la Cancelación

Tras el análisis estadístico (Chi-cuadrado) y la exploración visual, estos son los factores con mayor impacto:

A. Tiempo de Contrato
Es el predictor más fuerte. Existe una relación inversamente proporcional: cuanto menos tiempo lleva un cliente, más probable es que se vaya.

Punto crítico: La mayoría de las cancelaciones ocurren antes de los 12 meses. Una vez que el cliente supera los 24 meses, la probabilidad de fuga cae drásticamente.

B. Valor Mensual
El costo percibido es un disparador directo.

Los clientes que cancelan pagan, en promedio, facturas más altas (mediana cercana a 80 unidades) que los que permanecen (mediana cercana a 65). Los planes de alto costo sin una vinculación a largo plazo son los más vulnerables.

C. Estabilidad y Perfil del Cliente
Pareja y Dependientes: Los clientes que no tienen pareja o dependientes (perfiles más jóvenes o móviles) tienen tasas de cancelación significativamente más altas (33%) comparadas con perfiles familiares (20%).

Servicios de Valor Añadido: La ausencia de servicios como Seguridad Online o Soporte Técnico correlaciona fuertemente con la cancelación.

###Análisis del Rendimiento de los Modelos de Predicción


En esta etapa, comparamos dos enfoques distintos para predecir la fuga de clientes. La elección del modelo no se basó solo en quién acertaba más en general, sino en quién era capaz de detectar el problema real: el cliente que se va.

- El Modelo de Random Forest (Enfoque de Precisión Global)
Este modelo alcanzó una exactitud aparente del 79%. Sin embargo, descubrimos que este número es engañoso. Al ser un modelo "sin balancear", se volvió muy eficiente para identificar a los clientes que se quedan, pero falló críticamente en detectar a los que se van, logrando identificar solo al 48% de ellos (Recall). Es un modelo que "juega a lo seguro" y asume que la mayoría de los clientes son leales, ignorando a la mitad de los desertores.

- El Modelo de Regresión Logística (Enfoque Estratégico Balanceado)
Para solucionar la ceguera del modelo anterior, implementamos una Regresión Logística con balanceo de clases. Aunque su exactitud general bajó ligeramente al 74%, su efectividad real para el negocio se disparó:

  Capacidad de Detección (Recall): Logró identificar al 79% de los clientes que realmente cancelaron el servicio.

  Balance F1-Score: Obtuvo un 0.62, superando el 0.55 del modelo anterior, lo que demuestra un equilibrio mucho más sano entre precisión y detección.

--------------------------------------------------

Conclusión del modelo: El modelo de Regresión Logística con balanceo de clases es el ganador para el negocio. Aunque su exactitud general es menor, es capaz de identificar al 79% de los clientes que realmente se van, mientras que el Random Forest solo detecta a la mitad.

----------------------------------------------------

###Estrategias de Retención Propuestas y Recomendación


Basándonos en los resultados, se proponen las siguientes estrategias:

- Programa de "Fidelización Temprana": Dado que el riesgo es máximo en el primer año, se deben ofrecer descuentos decrecientes o beneficios de lealtad que se activen a partir del mes 6 y 12 para incentivar la permanencia inicial.

- Ajuste de Valor en Planes Premium: Identificar a los clientes que pagan más de 80 mensuales y ofrecerles servicios de valor añadido (Seguridad Online o Streaming) sin costo adicional para justificar el precio y aumentar la "viscosidad" del servicio.

- Incentivo a Contratos de Largo Plazo: Crear ofertas atractivas para migrar a clientes de "Mes a Mes" hacia contratos de uno o dos años, ya que la estabilidad del contrato reduce el Churn según la matriz de correlación.

- Campañas para Segmentos de "Baja Estabilidad": Diseñar planes específicos para clientes solteros o jóvenes, que incluyan beneficios de movilidad o flexibilidad que se adapten a su estilo de vida, reduciendo así su tendencia a la rotación.

- Se recomienda integrar el modelo de Regresión Logística en el sistema de atención al cliente para que, cuando un cliente de "alto riesgo" llame, el agente reciba una alerta automática con una oferta de retención personalizada basada en su perfil.
